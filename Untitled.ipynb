```python
import pandas as pd
example_yolo_log = '/Users/kpierce/tacc_transportation/lamar_rundberg/out.log'
example_yolo_log_columns = [
    'frame_idx',
    'object_trajectory_idx',
    'unclear_1',
    'unclear_2',
    'class_label',
    'class_confidence',
    'x_min',
    'y_min',
    'x_max',
    'y_max',
    'dx',
    'dy',
]
seq_dets = pd.read_csv(example_yolo_log)
seq_dets.columns = example_yolo_log_columns
ped_dets = seq_dets[(seq_dets['class_label'] == 0) & (seq_dets['class_confidence'] > 50)]
```

```python
total_time = 0.0
total_frames = 0

# create instance of the SORT tracker
mot_tracker = Sort(
    max_age=30,
    min_hits=3, # default
    iou_threshold=0.1
) 

tracks = []

for f in range(1, int(ped_dets['frame_idx'].max())):
        
    frame_dets = ped_dets[ped_dets['frame_idx'] == f]
    #print(frame_dets.head())

    dets = frame_dets[['x_min', 'y_min', 'x_max', 'y_max']].values
    #print(dets)
    total_frames += 1

    start_time = time.time()
    trackers = mot_tracker.update(dets)
    cycle_time = time.time() - start_time
    total_time += cycle_time
    
    for t in trackers.tolist():
        tracks.append(
            {'frame_idx': f, 'x_min': t[0], 'y_min': t[1], 'x_max': t[2], 'y_max': t[3], 'track_idx': t[4]}
        )
```

```python
tracks_df = pd.DataFrame.from_dict(tracks)
tracks_df['x_center'] = [(row['x_max'] + row['x_min'])/2 for i, row in tracks_df.iterrows()]
tracks_df['y_center'] = [(row['y_max'] + row['y_min'])/2 for i, row in tracks_df.iterrows()]
tracks_df.head()
```
